# Smart Study Analyser# 

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_excel("data/study_data.csv.xlsx")
df.head()
```

```python
df["Date"]= pd.to_datetime(df["Date"])
```

```python
df.info()
df.isnull().sum()
```

```python
time_analysis = ( df.groupby("Time_of_day")
["Test_score"].mean())
```

```python
time_analysis.plot(kind="bar",
title="Average Score by study Time")
plt.show()
```

```python
df["efficiency"] =  df["Test_score"] / df["Study_hour"]
```

```python
efficiency_analysis = df.groupby("Subject")["efficiency"].mean()
```

```python
daily_progress = df.groupby("Date")["Test_score"].mean()
daily_progress.plot(title="progress over Time")
sns.set()

```

```python
sns.barplot(x="Time_of_day", y ="Test_score" , data=df)
```

```python
sns.boxplot(x="Subject" ,y="Test_score" , data=df)
```

```python
sns.scatterplot(x="Study_hour", y ="Test_score",hue="Subject",data=df)
```
